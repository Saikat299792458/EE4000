# Sinusoidal data and its eigen value

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (15, 5)
M = 30;    # window length = embedding dimension
N = 200;   # length of generated time series
T = 22;    # period length of sine function
```

```python
t = np.linspace(1, N, N)
X = np.sin(2*np.pi*t/T);
plt.title("Time Series X")
plt.plot(t, X, 'b-');
```

### Calculate Covariance Matrix

```python
Y = np.zeros((N-M+1,M));
for m in range(M):
    Y[:, m] = X[m:N-M+m+1];
C = Y.T @ Y / (N-M+1);


fig, ax = plt.subplots()
plt.title('Covariance matrix');
plt.imshow(C);
#axis square
plt.clim(-1, 1); # Set Color limit of colorbar
plt.colorbar(); # Add colorbar
plt.show()
```

### Eigen value representation

```python
LAMBDA,RHO = np.linalg.eig(C);
SI = np.flip(np.argsort(LAMBDA))
LAMBDA = LAMBDA[SI]
RHO = RHO[:, SI];


plt.title('Eigenvectors RHO and eigenvalues LAMBDA')
plt.subplot(2,1,1);
plt.plot(LAMBDA,'o-');
plt.subplot(2,1,2);
plt.plot(RHO[:,0], '-', label='1');
plt.plot(RHO[:,1], '-', label='2');
plt.legend();
plt.tight_layout()
```

# 3 sinusoidal sum and its eigenvalue

```python
T1, T2, T3 = 10, 20, 30
X1 = np.sin(2*np.pi*t/T1);
X2 = np.sin(2*np.pi*t/T2);
X3 = np.sin(2*np.pi*t/T3);
plt.title("Time Series X")
plt.plot(t, X1, 'r-');
plt.plot(t, X2, 'g-');
plt.plot(t, X3, 'b-');
```

```python
XT = (X1 + X2 + X3) / 3
plt.plot(t, XT, 'b-');
plt.title("Sum of three sinusiods");
```

### Calculate Covariance Matrix

```python
Y = np.zeros((N-M+1,M));
for m in range(M):
    Y[:, m] = XT[m:N-M+m+1];
C = Y.T @ Y / (N-M+1);


fig, ax = plt.subplots()
plt.title('Covariance matrix');
plt.imshow(C);
#axis square
plt.clim(-1, 1); # Set Color limit of colorbar
plt.colorbar(); # Add colorbar
plt.show()
```

### Eigen Value representation

```python
LAMBDA,RHO = np.linalg.eig(C);
SI = np.flip(np.argsort(LAMBDA))
LAMBDA = LAMBDA[SI]
RHO = RHO[:, SI];

plt.figure(figsize = (15,10))
plt.title('Eigenvectors RHO and eigenvalues LAMBDA')
plt.subplot(4,1,1);
plt.plot(LAMBDA,'o-');
plt.subplot(4,1,2);
plt.plot(RHO[:,0], '-', label='1');
plt.plot(RHO[:,1], '-', label='2');
plt.legend();
plt.subplot(4,1,3);
plt.plot(RHO[:,2], '-', label='3');
plt.plot(RHO[:,3], '-', label='4');
plt.legend();
plt.subplot(4,1,4);
plt.plot(RHO[:,4], '-', label='5');
plt.plot(RHO[:,5], '-', label='6');
plt.legend();
plt.tight_layout()
```
